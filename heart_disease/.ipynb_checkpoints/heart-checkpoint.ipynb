## Load dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X, y = df.drop('target', axis=1), df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=9)
```

## Scalr-Insensitive

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf= GradientBoostingClassifier()
gb_clf.fit(X_train, y_train)
```

## Scale-Sensitive

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test) ## train
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn =  KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)

```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(X_train_scaled, y_train)
```

```python
from sklearn.svm import SVC #(support vector classifier)

svc =  SVC()
svc.fit(X_train_scaled, y_train)
```

```python
forest.score(X_test, y_test)
```

```python
 gb_clf.score(X_test, y_test)
```

```python
nb_clf.score(X_test, y_test)
```

```python
knn.score(X_test_scaled, y_test)
```

```python
log.score(X_test_scaled, y_test)
```

```python
svc.score(X_test_scaled, y_test)
```

```python
from sklearn.metrics import recall_score

y_preds = forest.predict(X_test)
print('forest:', recall_score(y_test, y_preds))

y_preds = nb_clf.predict(X_test)
print('GaussianNB:', recall_score(y_test, y_preds))

y_preds = gb_clf.predict(X_test)
print('GradientBoosting:', recall_score(y_test, y_preds))

y_preds = knn.predict(X_test_scaled)
print('KNeighbors:', recall_score(y_test, y_preds))

y_preds = log.predict(X_test_scaled)
print('logistic:', recall_score(y_test, y_preds))

y_preds = svc.predict(X_test_scaled)
print('SVC:', recall_score(y_test, y_preds))

```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100,200,300],
    'max_depth':[None,10,20,30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf':[1,2,4],
    'max_features':['sqrt','log2',None],
    
}

forest = RandomForestClassifier(n_jobs=1, random_state=9)
grid_search = GridSearchCV(forest, param_grid, cv=3, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest
```

```python
best_forest.score(X_test, y_test)
```

## Feature Importances

```python
import numpy as np

feature_importances = best_forest.feature_importances_
features = best_forest.feature_names_in_

sorted_idx = np.argsort(feature_importances)
sorted_features = features[sorted_idx]
sorted_importances = feature_importances[sorted_idx]

colors = plt.cm.YlGn(sorted_importances / max(sorted_importances))

plt.barh(sorted_features,sorted_importances, color=colors)
plt.xlabel('Features Importances')
plt.ylabel('Features')
plt.title('Feature Importances')
plt.show()
```

```python
import joblib
import os

# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
# Ensure these variables exist from previous cells
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
# scaler: The scaler object fitted on X_train
# grid_search: Your fitted GridSearchCV object (or 'forest', 'knn', etc.)
# feature_names: The list of feature names in the correct order
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---

# Define filenames
model_filename = 'heart_disease_model.joblib'
scaler_filename = 'heart_disease_scaler.joblib'
features_filename = 'heart_disease_features.joblib' # To store the feature order

# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
# Get the best model (adjust if not using GridSearchCV)
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
try:
    best_model = grid_search.best_estimator_
    print(f"Using best model from GridSearchCV: {best_model}")
except NameError:
    # Fallback if grid_search doesn't exist (check your variable names)
    try:
        best_model = forest # Or knn, etc.
        print(f"Using model variable: {best_model}")
    except NameError:
        print("ERROR: Could not find the trained model variable (e.g., grid_search, forest). Cannot save.")
        best_model = None

# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
# Get the scaler
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
try:
    scaler
    print(f"Using scaler: {scaler}")
except NameError:
    print("ERROR: Could not find the scaler variable. Cannot save.")
    scaler = None

# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
# Get feature names
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
try:
    # Try getting from X_train DataFrame columns if available
    feature_names = X_train.columns.tolist()
    print(f"Using feature names from X_train: {feature_names}")
except AttributeError:
    # Fallback: Define manually if X_train wasn't a DataFrame or columns aren't accessible
    # IMPORTANT: Ensure this order matches your training data EXACTLY
    feature_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
    print(f"Using manually defined feature names: {feature_names}")
except NameError:
     print("ERROR: Could not find X_train to get feature names. Using manual list.")
     feature_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']


# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
# Save the objects using joblib
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---
save_dir = 'saved_model' # Create a sub-directory
os.makedirs(save_dir, exist_ok=True)

model_path = os.path.join(save_dir, model_filename)
scaler_path = os.path.join(save_dir, scaler_filename)
features_path = os.path.join(save_dir, features_filename)

if best_model is not None:
    joblib.dump(best_model, model_path)
    print(f"Model saved to: {model_path}")
else:
    print("Model not saved because it wasn't found.")

if scaler is not None:
    joblib.dump(scaler, scaler_path)
    print(f"Scaler saved to: {scaler_path}")
else:
    print("Scaler not saved because it wasn't found.")

if feature_names:
    joblib.dump(feature_names, features_path)
    print(f"Feature names saved to: {features_path}")
else:
    print("Feature names not saved.")

```
